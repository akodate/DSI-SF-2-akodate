# Visualization of bootstrapping

The process of "bootstrapping" is resampling your data with replacement and calculating a statistic (or multiple statistics) of interest on those random samples. 

By doing this process, you achieve a _nonparametric_ method for calculating the confidence intervals around that statistic.

Why is that good? Because using parametric methods to caclulate statistics such as a t-test use assumptions about the distribution of your data, such as normality. While this is often true, you will encounter data where parametric methods are not appropriate.

---

## The purpose of the bootstrap

The bootstrap, and why it is effective, is based on the idea that the sample population that you have is representative of the overall population. This is the assumption you are making in all statistics for making an inference about the overall population.

Let's say we have a statistic of interest that we want to estimate. We also have a sample of observations drawn from the overall population that we are going to derive the statistic from. 

When this statistic is the mean, for example, then we know that with enough observations the distribution of sample means will be normal and the standard deviation of those means is derived from that distribution. We can then use that standard deviation to calculate the confidence intervals. 

Using the central limit theorem, we can say the calculation of a 95% confidence interval for our sample mean $\bar{X}$ is:

### $$ \bar{X} = 1.96 \frac{\sigma}{\sqrt{n}}$$

But let's say instead we want to get the confidence interval of the **_median_**. In this case, it is much harder to approximate the standard deviation of the overall population from the data. Bootstrapping allows us to bypass these assumptions entirely, making the calculation of the 95% confidence interval of the median rely only on our data.

---

###  How the bootstrap procedure works

Here is example code for bootstrapping the median of a sample of data `X`:

```python
X = sample_data
X_length = len(X)

bootstrap_iterations = 1000

bootstrap_sample_medians = []

for i in range(bootstrap_iterations):
    random_indicies = np.random.choice(range(X_length), size=X_length, replace=True)
    X_random_sample = X[random_indices]
    bootstrapped_median = np.median(X_random_sample)
    bootstrap_sample_medians.append(bootstrapped_median)
    
average_median = np.mean(bootstrap_sample_medians)
median_5th_pctl = np.percentile(bootstrap_sample_medians, 5)
median_95th_pctl = np.percentile(bootstrap_sample_medians, 95)

```

**Note: this is sampling _with replacement_.**

As you iterate in your for-loop the amount of times specified (in this case, 1000), you select with replacement a random sample of your original data `X`. This is because you are assuming that, since this data is representative of the overall population, you would get values "like this" if you sampled again, but not exactly the same. You assign an equal probability of pulling any of these values in your simulated new sample of data.

### Visual demonstrations of the bootstrap

Below are two different visualizations that demonstrate two scenarios where the bootstrap is extremely useful.

**1. Bootstrapping the confidence intervals of the median.**

The first chart shows the process, over iterations, of bootstrapping the confidence of the median. This distribution is specifically generated to be non-normal â€“ the bootstrap is not affected by strange distributions because it does not require assumptions about the distribution to work!

On the left is the histogram of your actual data, with the bootstrapped median and 95% confidence intervals around it. On the right is the distribution of the bootstrapped medians.

**2. Bootstrapping the confidence intervals of regression parameters.**

Bootstrapping can easily be extended to more complicated statistical measures such as a linear model. With a linear regression, we are fitting an intercept and coefficients on our variables (in this case, I am just showing the affect of one x variable on a y target variable. 

These beta coefficients are statistics just like the measurement of the median above. We can use the bootstrap procedure to figure out the confidence intervals around those coefficients, and then plot out the regression lines that represent the confidence interval space of where the regression line would fall. 

On the left chart there is the mean bootstrapped regression line plotted through the points, as well as the 5th and 95th percentile regression lines. The right chart shows the distribution of R2 values across the bootstrap samples.
