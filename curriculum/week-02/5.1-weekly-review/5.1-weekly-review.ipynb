```python
# Group the titanic data by Pclass
# normalize fair for each group, and assign the normalized score a new
# column 'Fare_norm'
# plot the grouped boxplot again, like above, with seaborn.
```

```python
def normalize(df):
    df['fare_norm'] = (df['Fare'] - df['Fare'].mean())/df['Fare'].std()
    return df

titanic = titanic.groupby('Pclass').apply(normalize)
```

```python
def kiefers_groupby(df, column):
    
    groups = []
    
    unique_elements_in_col = df[column].unique()
    #print column, unique_elements_in_col
    
    for unique_element in unique_elements_in_col:
        
        subset = df.loc[(df[column] == unique_element), :]
        
        #print subset.head(2)
        #print '--------------------------------------\n'
        
        groups.append([unique_element, subset])
        
    return groups

groups = kiefers_groupby(titanic, 'Pclass')
print len(groups)
```

```python
def kiefers_apply(groups, function):
    
    modified_subsets = []
    
    for group in groups:
        modified = function(group[1])
        modified_subsets.append(modified)
        
    recombined = pd.concat(modified_subsets, axis=0)
    return recombined
    
    
```

```python
def some_function(df):
    df.loc[:, 'num_rows'] = df.shape[0]
    return df
```

```python
newdf = kiefers_apply(groups, some_function)
newdf.num_rows.unique()
```
