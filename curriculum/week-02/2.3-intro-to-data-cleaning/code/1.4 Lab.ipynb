```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
rock_filepath = '/Users/alex/Desktop/DSI-SF-2-akodate/curriculum/week-02/2.4-lab-cleaning-lab/assets/datasets/rock.csv'

rock_songs = pd.read_csv(rock_filepath)
```

```python
print rock_songs.info()
print rock_songs['First?'].value_counts()
rock_songs
```

```python
rock = rock_songs.drop(['First?', 'Year?'], axis=1)
```

```python
rock = rock.rename(columns={
        'Song Clean': 'song_name',
        'ARTIST CLEAN': 'artist_name',
        'Release Year': 'release_year',
        'COMBINED': 'song_and_artist',
        'PlayCount': 'play_count',
        'F*G': 'year_times_play_count'
    })

rock.head()
```

```python
rock['release_year'] = pd.to_numeric(rock['release_year'], errors='coerce')
rock['release_year'] = rock[(rock['release_year'] > 1900) | (rock['release_year'].isnull())]['release_year']
```
